# Visualizing Categorical Variables (Part 2)

Similar to the previous step, it creates pie charts for each categorical variable's distribution.

```python
num_cat_vars = len(cat_var)
num_cols = 3  # You can adjust the number of columns as needed
num_rows = (num_cat_vars + num_cols - 1) // num_cols

# Creating Fig with subplot
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(20, 10))
axs = axs.flatten()

# Creating countplot for each categorical variable
for i, var in enumerate(cat_var):
    sns.countplot(x=var, data=dataset, hue = 'status', ax=axs[i])
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=90)

# Remove the empty subplot in the last position
# Remove the empty subplot in the last position
if len(axs) > num_cat_vars:
    fig.delaxes(axs[-1])
    fig.delaxes(axs[-2])
# Adjust spacing between subplots
fig.tight_layout()

# Show plot
plt.show()
```

```python
# Create a grid of subplots based on the number of categorical variables
num_cat_vars = len(cat_var)
num_cols = 3  # You can adjust the number of columns as needed
num_rows = (num_cat_vars + num_cols - 1) // num_cols

fig, axs = plt.subplots(num_rows, num_cols, figsize=(10, 10))
axs = axs.flatten()  # Flatten the 2D array of axes for easier indexing

# Loop through categorical variables and create countplots
for i, var in enumerate(cat_var):
    if i < len(axs):
        ax = axs[i]
        cat_counts = dataset[var].value_counts()
        ax.pie(cat_counts, labels=cat_counts.index, autopct='%1.1f%%', startangle=90)
        ax.set_title(f'{var} Distribution')
    
# Adjust spacing between subplots
fig.tight_layout()

# If there's an extra subplot that needs to be removed
if len(axs) > num_cat_vars:
    fig.delaxes(axs[-1])
    fig.delaxes(axs[-2])

# Show the plot
plt.show()
```

# Data Preprocessing (Label Encoding and One-Hot Encoding)

It uses label encoding to convert the 'status' column (placement status) into numerical values (0 or 1).

It then performs one-hot encoding on categorical variables to convert them into binary (0 or 1) columns.

```python
le = LabelEncoder()
dataset['status']=le.fit_transform(dataset['status'])
```

```python
y = dataset[['status']]
```

```python
gender = pd.get_dummies(dataset[['gender']],drop_first=True)
ssc_board = pd.get_dummies(dataset[['ssc_board']],drop_first=True)
hsc_board = pd.get_dummies(dataset[['hsc_board']],drop_first=True)
hsc_subject = pd.get_dummies(dataset[['hsc_subject']],drop_first=True)
undergrad_degree = pd.get_dummies(dataset[['undergrad_degree']],drop_first=True)
work_experience = pd.get_dummies(dataset[['work_experience']],drop_first=True)
specialisation = pd.get_dummies(dataset[['specialisation']],drop_first=True)
```

```python
#Drop the previous catagorical variable
cat_var = ['gender', 'ssc_board', 'hsc_board',
           'hsc_subject', 'undergrad_degree',
           'work_experience', 'specialisation',
            'status'
]
for var in cat_var: 
    dataset.drop(columns=var,inplace=True)
```

```python
dataset.head()
```

```python
#concat dummy variable to the original dataset
dataset = pd.concat([
    dataset, 
    gender,
    ssc_board,
    hsc_board,
    hsc_subject,
    undergrad_degree,
    work_experience,
    specialisation,
], axis=1)
```

# Splitting the Dataset

It splits the dataset into training and testing sets using train_test_split.

```python
x_train,x_test,y_train,y_test=train_test_split( x, y, test_size=0.20, random_state=4)
```

# Building and Training a K-Nearest Neighbors Classifier

It creates a K-Nearest Neighbors (KNN) classifier model.

It fits the model on the training data.

```python
model = KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

# Making Predictions and Evaluating the Model

It uses the trained model to make predictions on the testing data.

It calculates a confusion matrix, which shows the true positive, true negative, false positive, and false negative values.

It visualizes the confusion matrix using Seaborn.

It calculates and prints accuracy, precision, recall, and F1-score as evaluation metrics for the model.

```python
y_pred = model.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
import numpy as np

# Create a confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['True 0', 'True 1'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
# Calculate precision score
precision = precision_score(y_test, y_pred)
print("Precision Score:", precision)
```

```python
# Calculate recall score
recall = recall_score(y_test, y_pred)
print("Recall Score:", recall)
```

```python
# Calculate F1 score
f1 = f1_score(y_test, y_pred)

print("F1 Score:", f1)
```
